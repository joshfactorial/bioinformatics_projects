```python
def count_letters(sequence: str) -> list:
    letters = ['A', 'C', 'G', 'T']
    counts = []
    for character in letters:
        count = 0
        for i in range(len(sequence)):
            if sequence[i] == character:
                count += 1
        counts.append(count)
    return counts
```

```python
def replace_t_with_u(sequence: str) -> str:
    RNA = ""
    for i in range(len(sequence)):
        if sequence[i] == 'T':
            RNA = RNA + 'U'
        else:
            RNA = RNA + sequence[i]
    return RNA
```

```python
def rev_comp(sequence: str) -> str:
    comps = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
    new_seq = ""
    for i in range(len(sequence)):
        new_seq = comps[sequence[i]] + new_seq
    return new_seq
```

```python
def rabbits_after_months(n: int, k: int) -> int:
    F_n = 0
    F_n1 = 1
    F_n2 = 1
    for i in range(n-2):
        F_n = F_n1 + (k * F_n2)
        F_n2 = F_n1
        F_n1 = F_n
    return F_n
```

```python
n = 33
k = 3
answer = rabbits_after_months(n, k)
print(answer)
```

```python
def get_gc_content(sequence: str) -> float:
    count = 0
    for letter in sequence:
        if letter == "G" or letter == "C":
            count += 1
    return count/len(sequence)

def read_fasta(filename: str) -> dict:
    dictionary = {}
    temp_key = ""
    temp_sequence = ""
    first_time = 1
    file = open(filename, 'r')
    
    for line in file:
        line = line.rstrip()
        if line[0] == ">":
            if first_time == 1:
                first_time = 0
                temp_key = line[1:]
            else:
                dictionary[temp_key] = temp_sequence
                temp_sequence = ""
                temp_key = line[1:]
        else:
            temp_sequence += line
    dictionary[temp_key] = temp_sequence
    return dictionary

def process_fasta_gc(dictionary: dict) -> list:
    gc_percents = []
    gc_keys = []
    for key in dictionary:
        gc_keys.append(key)
        gc_percents.append(get_gc_content(dictionary[key]))
    index = gc_percents.index(max(gc_percents))
    return gc_keys[index], gc_percents[index]
```

```python
def hamming_distance(string1: str, string2: str) -> int:
    distance = 0
    if len(string1) != len(string2):
        print("Can't compare these two strings")
    else:
        for i in range(len(string1)):
            if string1[i] != string2[i]:
                distance += 1
    return distance        
```
