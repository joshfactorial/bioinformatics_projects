```python
def count_letters(sequence: str) -> list:
    letters = ['A', 'C', 'G', 'T']
    counts = []
    for character in letters:
        count = 0
        for i in range(len(sequence)):
            if sequence[i] == character:
                count += 1
        counts.append(count)
    return counts
```

```python
def replace_t_with_u(sequence: str) -> str:
    RNA = ""
    for i in range(len(sequence)):
        if sequence[i] == 'T':
            RNA = RNA + 'U'
        else:
            RNA = RNA + sequence[i]
    return RNA
```

```python
def rev_comp(sequence: str) -> str:
    comps = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
    new_seq = ""
    for i in range(len(sequence)):
        new_seq = comps[sequence[i]] + new_seq
    return new_seq
```

```python
sequence = "AAAACCCGGT"
answer = rev_comp(sequence)
print(answer)
```

```python
def rabbits_after_months(n: int, k: int) -> int:
    F_n = 0
    F_n1 = k
    F_n2 = 1
    for i in range(n):
        F_n = k * F_n1 + F_n2
        F_n2 = F_n1
        F_n1 = F_n
    return F_n
```

```python
n = 5
k = 3
answer = rabbits_after_months(n, k)
print(answer)
```
